```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Datos

```python
# Cargar el CSV con una codificación específica
df = pd.read_csv('superstore.csv', encoding='ISO-8859-1')

# Visualizar las primeras filas del dataset
df.head()
```

## 2. Exploración y Limpieza de los Datos

```python
print(df.describe())
```

```python
# vemos si es que existen valores perdidos
print(df.isnull().sum())
```

## 3. Análisis Exploratorio de Datos (EDA)

### a) Ventas por Categoría de Producto

```python
# Ventas por categoría
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Sales', data=df, estimator=sum)
plt.title('Total de Ventas por Categoría')
plt.show()
```

### b) Ventas por Región

```python
# Ventas por región
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Sales', data=df, estimator=sum)
plt.title('Total de Ventas por Región')
plt.show()
```

### c) Ventas por Segmento de Cliente

```python
# Ventas por segmento de cliente
plt.figure(figsize=(10, 6))
sns.barplot(x='Segment', y='Sales', data=df, estimator=sum)
plt.title('Total de Ventas por Segmento de Cliente')
plt.show()
```

### d) Relación entre Descuento y Ventas

```python
# Relación entre descuento y ventas
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Sales', data=df)
plt.title('Relación entre Descuento y Ventas')
plt.show()
```
